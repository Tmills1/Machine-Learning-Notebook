# Machine Learning
Machine learning is an application of Artificial Intelligence that enables systems to learn and improve from experience without being explicitly programmed. Machine learning focuses on developing computer programs that can access data and use it to learn for themselves.

There are three major types of Machine Learning methods: `supervised, unsupervised and semi-supervised learning`.

### What is supervised learning?

This approach basically teaches machines by example. Supervised machine learning is a type of machine learning where an algorithm learns from labeled training data to make predictions or take actions. It involves providing the model with input data and corresponding correct output labels, allowing it to learn the mapping between the inputs and outputs.

During training for supervised learning, systems are exposed to large amounts of labelled data, for example images of handwritten figures annotated to indicate which number they correspond to. Given sufficient examples, a supervised-learning system would learn to recognize the clusters of pixels and shapes associated with each number and eventually be able to recognize handwritten numbers, able to reliably distinguish between the numbers 9 and 4 or 6 and 8.

However, training these systems typically requires huge amounts of labelled data, with some systems needing to be exposed to millions of examples to master a task.

### What is unsupervised learning?
In contrast, unsupervised learning tasks algorithms with identifying patterns in data, trying to spot similarities that split that data into categories.

An example might be Airbnb clustering together houses available to rent by neighborhood, or Google News grouping together stories on similar topics each day.Unsupervised learning algorithms aren't designed to single out specific types of data, they simply look for data that can be grouped by similarities, or for anomalies that stand out.

### What is semi-supervised learning?
As the name suggests, the approach mixes supervised and unsupervised learning. The technique relies upon using a small amount of labelled data and a large amount of unlabelled data to train systems. The labelled data is used to partially train a machine-learning model, and then that partially trained model is used to label the unlabelled data, a process called pseudo-labelling. The model is then trained on the resulting mix of the labelled and pseudo-labelled data

## How does supervised machine learning work?

With supervised learning, the model is trained until it can detect relationships and patterns between the input and output variables. It's ideal for classification and regression problems:
- **Classification**: The target variable in classification is categorical, consisting of discrete classes or categories. The goal is to assign the data points to the correct class based on the input features.

- **Regression**: In regression, the target variable is continuous and numeric. The objective is to predict a value that lies on a continuous scale. For example, predicting house prices based on features like area, number of rooms, etc.

# REGRESSION ANALYSIS

Before we delve into the details, there are some of the key terminologies used in linear regression. Understanding these terms is essential for interpreting and analyzing the results of a linear regression model.

- **Dependent variable:** Also known as the response variable or the target variable, it is the variable that we want to predict or explain using the independent variables. It is denoted as "y" in the context of linear regression.

- **Independent variables:** Also known as predictor variables or features, these are the variables used to predict or explain the dependent variable. In linear regression, there can be one or more independent variables. They are denoted as "X" or "x" in the context of linear regression.

- **Coefficients**: Coefficients, also called weights or parameters, represent the estimated values in the linear regression model. They indicate the strength and direction of the relationship between the independent variables and the dependent variable. In a simple linear regression model (with one independent variable), there is a single coefficient. In multiple linear regression (with multiple independent variables), there is a coefficient corresponding to each independent variable.

- **Intercept:** The intercept, denoted as `b0` or `β0`, represents the expected value of the dependent variable when all the independent variables are set to zero. It represents the baseline or starting point of the regression line.

- **Residuals:** Residuals, also known as errors, are the differences between the actual values of the dependent variable and the predicted values by the linear regression model. They represent the unexplained variation in the dependent variable.

- **Ordinary Least Squares (OLS):** OLS is the most common method used to estimate the coefficients in linear regression. It `minimizes the sum of squared residuals to find the best-fitting line`.

- **R-squared:** R-squared (R²) is a statistical measure that represents the proportion of the variance in the dependent variable that is explained by the independent variables in the linear regression model. It ranges from 0 to 1, where 0 indicates that the model explains none of the variance, and 1 indicates a perfect fit.

- **Multicollinearity:** Multicollinearity occurs when two or more independent variables in the linear regression model are highly correlated with each other. It can lead to unstable or unreliable coefficient estimates and makes it challenging to interpret the individual effects of the predictors.

- **Heteroscedasticity:** Heteroscedasticity refers to a situation where the variance of the residuals is not constant across different values of the independent variables.

- **Outliers:** Outliers are extreme values that deviate significantly from the other data points. They can have a disproportionate impact on the regression line and the estimated coefficients. Identifying and addressing outliers is important in linear regression.

## Assumption of Linear Regression

Linear regression makes several assumptions about the underlying data and the relationship between the dependent variable and the independent variables. These assumptions are important to ensure that the linear regression model produces reliable and meaningful results. Here are the key assumptions of linear regression:

- **Linearity**: The relationship between the dependent variable and the independent variables is assumed to be linear. This means that the effect of the independent variables on the dependent variable is a constant and additive.

- **Independence**: The observations or data points used in the regression model are assumed to be independent of each other. There should be no systematic relationship or correlation between the residuals (the differences between the actual and predicted values) of the model.

- **Homoscedasticity**: Homoscedasticity assumes that the variance of the residuals is constant across all levels of the independent variables. In other words, the spread or dispersion of the residuals should be consistent for different values of the predictors.

- **Normality**: Linear regression assumes that the residuals follow a normal distribution. This is important for statistical inference and hypothesis testing. If the residuals are not normally distributed, it can affect the accuracy of confidence intervals and p-values.

- **No multicollinearity**: Multicollinearity occurs when the independent variables in the regression model are highly correlated with each other. It can make it difficult to interpret the individual effects of the predictors on the dependent variable.

- **No endogeneity**: Endogeneity refers to a situation where there is a bidirectional relationship between the dependent variable and one or more independent variables. This violates the assumption of causality in linear regression, where the independent variables should not be influenced by the dependent variable.

- **No outliers**: Outliers are extreme values that deviate significantly from the other data points. They can have a strong influence on the regression model and affect its coefficients and predictions. Linear regression assumes that there are no influential outliers in the data.

It is important to assess these assumptions when applying linear regression and take appropriate steps if any of the assumptions are violated.

### Train-Test Split

In machine learning, data splitting refers to the process of dividing a dataset into separate subsets for training and testing a machine learning model. The two main subsets commonly used are the train set and the test set. Here's an explanation of each set:

- **Training Set:** the portion of the dataset that is used to train or fit the machine learning model. It is the data on which the model learns the underlying patterns, relationships, and parameters from the available features and corresponding target values. The model adjusts its internal parameters based on the training set in order to minimize the error or loss function. It typically contains a majority portion of the available data, such as 70-80% of the dataset.

- **Test Set:**  a separate portion of the dataset that is used to evaluate the performance and generalization ability of the trained model. It serves as an unseen data sample to assess how well the model performs on new, unseen instances. The test set should be independent and representative of the data distribution that the model will face during its operational phase. It is  smaller than the training set, typically around 20-30% of the dataset.

Splitting the dataset allows the model to understand our datasets by fitting the training set. It also helps to measure the model's ability to generalize well on unseen data. By evaluating the model's performance on the test set, we can estimate how well it is likely to perform on new, unseen instances in the real world.

```python
def split_data(features, targets, test_size=0.2):
    # Split the data into training and test sets
    num_instances = len(features)
    num_test = int(test_size * num_instances)
    #shuffle the rows before splitting data
    indices = np.random.permutation(num_instances)
    train_indices = indices[num_test:]
    test_indices = indices[:num_test]
    train_features = features[train_indices]
    test_features = features[test_indices]
    train_targets = targets[train_indices]
    test_targets = targets[test_indices]
    # Return the training set and test set
    return train_features, test_features, train_targets, test_targets
```

### Preprocessing Data: Feature scaling

Feature scaling is a data preprocessing method commonly used in machine learning to standardize or normalize the features of a dataset. It involves transforming the feature values to a similar scale or range to ensure that no particular feature dominates the learning algorithm due to its larger magnitude. Feature scaling is particularly important in machine learning algorithms that are sensitive to the scale of features or when features have different units of measurement. Here are two common methods for feature scaling:

- **Min-Max Normalization:** Min-max scaling, also known as normalization, scales the feature values to a specific range, typically between 0 and 1. The formula for min-max scaling is:
  $$ \bar{X} = \frac{X - {X_{min}}}{{X_{max}} - {X_{min}}}$$

  This method preserves the relative relationships between the feature values and ensures that they fall within a specified range.
- **Standardization:** Standardization, also known as z-score scaling, transforms the feature values to have a mean of 0 and a standard deviation of 1. The formula for standardization is:
   $$ \bar{X} = \frac{X - {X_{mean}}}{{X_{std}}}$$

  Standardization centers the feature values around the mean and scales them based on the standard deviation. It makes the features more comparable by removing the differences in their means and scales.

The choice between min-max scaling and standardization depends on the specific characteristics of the dataset and the requirements of the learning algorithm. Min-max scaling is suitable when you want to preserve the original range of the data, while standardization is useful when you want to give equal importance to all features and when the algorithm assumes that the features are normally distributed.

It's important to note that feature scaling should be **applied separately to the training set and the test set**. The scaling parameters (e.g., `min-max values`, `mean`, `standard deviation`) **obtained from the training set should be used to scale the test set to ensure consistency**.

Now that you understand the basic concepts of feature scaling, we will implement a function called `scale_data`. This function will take in an array of independent variable(s) and scale them based on a specified method: `standardize` or `normalize`.

```python
def scale_data(train_features, test_features, scaling_method):
    # Preprocess the data (e.g., normalization, feature scaling, etc.)
    if scaling_method == "standardize":
        # Return preprocessed features
        standardized_train_features = (train_features - np.mean(train_features, axis=0)) / np.std(train_features, axis=0)
        #Use the statistics of the training set to scale the test set.
        standardized_test_features = (test_features - np.mean(train_features, axis=0)) / np.std(train_features, axis=0)
        return standardized_train_features, standardized_test_features
    elif scaling_method == "normalize":
        # Return preprocessed features
        normalized_train_features = (train_features - np.min(train_features, axis=0)) / (np.max(train_features, axis=0) - np.min(train_features, axis=0))
        #Use the statistics of the training set to scale the test set.
        normalized_test_features = (test_features - np.min(train_features, axis=0)) / (np.max(train_features, axis=0) - np.min(train_features, axis=0))
        return normalized_train_features, normalized_test_features
```

## Implementing a Simple Linear Regression Model

Let's start basic. We will implement a simple regression model. This model will make use of one predictor to estimate the value of house prices.

We'll start by understanding the univariate case of linear regression, also known as simple linear regression. The following equation is the general form of the simple linear regression model.

$$ y = {a_1}{x_1} + {b_0}$$

- $y$ represents the target column.

- ${x_1}$ represents the feature column we choose to use in our model.

- ${b_0}$ and ${a_1}$ represent the **parameter** values tfrom our model. ${a_1}$ is the coefficient associated with the predictor column ${x_1}$ (otherwise clalled `slope`). ${b_0}$ is called a bias term ( also called `intercept` in simple linear regression).


To find the coefficients, we need to estimate β₀ and β₁ using the given data points (x, y). The formulas for the coefficients in simple linear regression are:

$$ {a_0} = \frac{Σ(({x_i} - \bar{x})({y_i} -\bar{y}))}{Σ({x_i} - \bar{x})²)} $$

$${b_0} = \bar{y} - ({a_1} * \bar{x})$$

Here, (${x_i},{y_i}$) represents each data point, $\bar{x}$  is the mean of the independent variable $x$, and $\bar{y}$ is the mean of the dependent variable $y$.

For more clarity, see [here](https://youtu.be/m88h75F3Rl8).


The first step is to select the feature, ${x_1}$ we want to use in our model. Let's attempt to predict using the `CRIM` feature. Next, we will build a function called `parameters` that estimates the coefficient and bias terms.

```python
# Calculate covariance between x and y
def parameters(predictor, target):
  # Calculate the coefficients for simple linear regression
    predictor_mean = np.mean(predictor)
    target_mean = np.mean(target)
    numerator = np.sum
    numerator = np.sum((predictor - predictor_mean) * (target - target_mean))
    denominator = np.sum((predictor - predictor_mean) ** 2)
    slope = numerator / denominator
    intercept = target_mean - slope * predictor_mean
    return slope, intercept
```

We will then use the training set to obtain the model parameters using the `parameter` function.

```python
predictor = train_features[:,0] #CRIM feature
target = train_targets
slope, intercept = parameters(predictor, target)
print("Intercept:", intercept)
print("Slope:", slope)
```

### Evaluation Metrics for Linear Regression

In Machine Learning, an evaluation metric is used to determine and estimate the performance of a model.  The strength of any linear regression model can be assessed using various evaluation metrics. These evaluation metrics usually provide a measure of how well the observed outputs are being generated by the model. These metrics are used to compare the predicted values with the actual values of houses in the test set.

The most used metrics in regression are:

- **Coefficient of Determination or R-Squared (R2):** R-Squared is a number that explains the amount of variation that is explained/captured by the developed model. It always ranges between 0 & 1 . Overall, the higher the value of R-squared, the better the model fits the data.

  $$ {R_{squared}} = 1 - \frac{ssr}{sst}$$

  where `ssr` is the sum of squared residuals (sum of squared differences between the true and predicted values) and `sst` is the total sum of squares (sum of squared differences between the true values and their mean).

- **Mean Squared error (MSE)**: a commonly used metric for evaluating the performance of regression models. MSE measures the average squared difference between the predicted values and the true values in a regression problem.

$$ MSE = \frac{1}{n}\Sigma_{i=1}^{n}{\Big(\bar{y_i} - y_i\Big)^2} $$

- **Mean Absolute Error (MAE):** MAE measures the average absolute difference between the predicted values and the true values in a regression problem. Mathematically it can be represented as,

$$ MAE = \frac{1}{n}\Sigma_{i=1}^{n}{\left\lvert{\bar{y_i} - {y_i}}\right\rvert^2} $$


The closer the error (MSE or MAE) is to zero, the better the model.

Let's implement a function `evaluate_regression` that computes the R-squared score, mean squared error (MSE), and mean absolute error (MAE). The function `evaluate_regression()` will take two input arrays `y_true` and `y_pred`, representing the true target values and predicted target values, respectively.

```python
def evaluate_regression(y_true, y_pred):
    # Compute R-squared score
    ssr = np.sum((y_true - y_pred) ** 2)
    sst = np.sum((y_true - np.mean(y_true)) ** 2)
    r_squared = 1 - (ssr / sst)

    # Compute mean squared error (MSE)
    mse = np.mean((y_true - y_pred) ** 2)

    # Compute  mean absolute error (MAE)
    absolute_errors = np.abs(y_true - y_pred)
    mae = np.mean(absolute_errors)

    return r_squared, mse, mae
```

```python
r2_score, mse, mae = evaluate_regression(test_targets, predicted_prices)
print("R-squared Score:", r2_score)
print("MSE:", mse)
print("MAE:", mae)
```

## Overfitting and Underfitting

Underfitting and overfitting are two common issues that can occur when training machine learning models. These issues are related to the model's ability to generalize well to unseen data. Let's explore them further:

- **Underfitting**:
Underfitting occurs when a model is too simple to capture the underlying patterns and relationships in the data. It fails to learn from the training data effectively and performs poorly on both the training and test data. In terms of the Mean Squared Error (MSE) score, an underfit model would have a relatively high MSE on both the training and test sets. It indicates that the model is not able to sufficiently capture the complexity of the data, leading to substantial errors in predictions.

- **Overfitting**:
Overfitting occurs when a model becomes too complex and starts to memorize the noise or random fluctuations in the training data. It learns the training data too well and performs exceptionally well on the training set but fails to generalize to new, unseen data. In terms of the MSE score, an overfit model would have a very low MSE on the training set but a significantly higher MSE on the test set. This suggests that the model is fitting the noise and idiosyncrasies of the training data rather than the true underlying patterns.


image.png


Both underfitting and overfitting are undesirable as they indicate poor model performance. The goal is to find the right balance or complexity in the model that allows it to generalize well to unseen data without underfitting or overfitting

# Classification


Classification is a type of supervised learning in machine learning (ML) that involves categorizing or classifying data into predefined classes or categories based on their features. The objective of classification is to build a model that can accurately assign new, unseen data points to the appropriate class based on the patterns and relationships learned from the training data.

There are two main types of classification problems in ML:
- **Binary Classification:** In binary classification, the task is to classify data into one of two classes or categories. For example, determining whether an email is spam or not spam, predicting whether a customer will churn or not, or classifying an image as containing a specific object or not.

- **Multiclass Classification:** In multiclass classification, the task is to classify data into more than two classes or categories. Each data point is assigned to a single class out of multiple options. For instance, classifying an email into different categories (e.g., spam, social, promotions), recognizing handwritten digits (0-9), or identifying different types of animals.

The key distinction between classification and regression lies in the nature of the target variable.While both classification and regression are supervised learning tasks, they differ in terms of the nature of the target variable and the objective of the prediction. Classification focuses on assigning data points to predefined classes, while regression aims to estimate a numerical value based on the input features.

## KNearest Neighbors


K-Nearest Neighbors (KNN) is a simple and intuitive algorithm used for both classification and regression tasks in machine learning. It is an instance-based learning algorithm that makes predictions based on the similarity between new data points and the known data points in the training set.

In KNN, the "K" represents the number of nearest neighbors that are considered for making predictions. When a new data point needs to be classified or predicted, the algorithm calculates the distance between that point and all other points in the training set. The K nearest neighbors with the shortest distances to the new data point are then selected.

For classification tasks, the majority class among the K nearest neighbors is assigned to the new data point. In other words, the class that occurs most frequently among the neighbors is chosen as the predicted class for the new point.

image.png

KNN operates on the assumption that similar data points tend to belong to the same class or have similar target values. The choice of the value K is an important consideration in KNN. A smaller value of K can lead to more flexible decision boundaries but might be more sensitive to noise or outliers. A larger value of K can provide smoother decision boundaries but might overlook local patterns.

It is worth noting that KNN does not involve an explicit training phase as it relies on storing the entire training dataset for making predictions. This can make KNN computationally expensive for large datasets, as the distance calculations need to be performed for each new data point.

For more clarity on Knearest neighbours, see [intuitive explanation](https://youtu.be/0p0o5cmgLdE) and [KMM classifier](https://youtu.be/4HKqjENq9OU).

#### Implementing a KNearest Neighbors Classifier
In this section, we will implement a KNearest Neighbor Classifier.

### Euclidean Distance
Euclidean distance is a measure of the straight-line distance between two points in a Euclidean space. It is widely used in various fields, including mathematics, physics, and machine learning

image.png

Now we will implement a function called `euclidean_distance()` which takes two input arrays instance1 and instance2, for which we want to calculate the Euclidean distance.

```python
def euclidean_distance(instance1, instance2):
    """Calculate the Euclidean distance between two arrays"""
    distance = np.sqrt(np.sum((instance1 - instance2)**2))
    return distance
```

Let's put the euclidean distance function to a test.

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
distance = euclidean_distance(array1, array2)
print("Euclidean Distance:", distance)
```

Remember that KNearest Neighbors isn't actually trained on the dataset. It actually stores the training Data. Whn presented with an instance in the test set, the algorithm:
1.  searches through all instances in the training set.
2.  computes the distance between a pair of data points from the trainijng set and an instance in the test set.
3.  select neighbour: the algorithm selects `K` nearest data points in terms of euclidean distance to the data point from the test set.
4.  Checks the class that all `K` neighbors belong to and votes the majority class as prediction.

We will implement a function `get_k_nearest_neighbors()` that does steps 1-3. This function will take in the test instance, the training dataset and a specified `K` value. The function will then return the `K` nearest neighbors.

```python
def get_k_nearest_neighbors(test_instance, training_set,training_labels, k):
    # get distance between the test isntance and every data point in the training set.
    distances = [euclidean_distance(test_instance, train_instance) for train_instance in training_set]
    sorted_indices = np.argsort(distances)
    # Find the k nearest neighbors of the test instance
    k_nearest_neighbors = training_set[sorted_indices[:k]]
    k_nearest_labels = training_labels[sorted_indices[:k]]
    # Return the k nearest neighbors
    return k_nearest_neighbors,k_nearest_labels
```

Let's attempt to use this function to get the closest neighbor to the fifth patient in our dataset. Hence, we will use a `1` neighbors benchmark.

```python
test_case = X_test.values[4,:] #5th patient in our test set
k_nearest_neighbors,k_nearest_labels = get_k_nearest_neighbors(test_case, X_train.values,y_train.values, k=1)
k_nearest_neighbors,k_nearest_labels
```

- This gives an array containing the attributes of the closest patient.

Remember step 4?
The algorithm is expected to check the class that all `K` neighbors belong to and vote the majority class as prediction. We will implement a function called `predict` which will achieve step 4. This function will take in all the parameters that `get_k_nearest_neighbors()` also takes in. Additionally, it will take in the labels of datapoints in the training set.

```python
from collections import Counter
from scipy.stats import mode

def predict(test_instance, training_set, labels, k):
    """Predict the class label for a test instance"""
    neighbors,labels = get_k_nearest_neighbors(test_instance, training_set,labels, k)
    # Return the predicted class label
    most_common_label = mode(labels).mode[0]
    return most_common_label
```

Let's use this function to predict whether the fifth patient in our test set is at risk of a heart disease. We will specify a 5 nearest neighbor's model.

## Evaluation Metrics for Classification

Evaluation metrics for classification tasks assess the performance and effectiveness of a classification model by comparing the predicted class labels with the true class labels. Here are some commonly used evaluation metrics for classification:

- **Accuracy:** Accuracy measures the proportion of correctly classified instances out of the total number of instances. It provides an overall measure of how well the model predicts the correct class labels. However, accuracy can be misleading if the classes are imbalanced.

- **Precision:** Precision calculates the proportion of correctly predicted positive instances (true positives) out of the total instances predicted as positive (true positives + false positives). Precision focuses on the accuracy of positive predictions and is useful when the cost of false positives is high.

- **Recall (Sensitivity or True Positive Rate):** Recall measures the proportion of correctly predicted positive instances (true positives) out of the total actual positive instances (true positives + false negatives). Recall focuses on capturing as many positive instances as possible and is valuable when the cost of false negatives is high.

- **F1-Score:** The F1-score is the harmonic mean of precision and recall, providing a balanced measure that considers both precision and recall. It combines the precision and recall into a single metric, useful when both false positives and false negatives are equally important.

However, we will only implement accuracy in this section. The other metrics can be called from sklearn's [metrics](https://scikit-learn.org/stable/modules/model_evaluation.html) module.

```python
def evaluate(predictions, actual_labels):
    """Evaluate the performance of the k-NN classifier using accuracy"""
    correct_predictions = np.sum(predictions == actual_labels)
    #compute the accuracy
    accuracy = correct_predictions / len(actual_labels)
    return accuracy
```

Let's tie it all together!

```python
def knn_classifier(train_features, test_features, test_labels, train_labels, k):
    """Main function to build and evaluate the k-NN classifier"""
    predicted_labels = []
    for test_instance in test_features:
        predicted_label = predict(test_instance, train_features, train_labels, k)
        predicted_labels.append(predicted_label)
    print(len(predicted_labels))
    evaluation_metrics = evaluate(predicted_labels, test_labels)
    print(f"Accuracy on test dataset: {evaluation_metrics}")
    return predicted_labels
```
